```python
df['Diastolic blood pressure (mm Hg)']=df['Diastolic blood pressure (mm Hg)'].replace(0,df['Diastolic blood pressure (mm Hg)'].mean())
df['Triceps skin fold thickness (mm)']=df['Triceps skin fold thickness (mm)'].replace(0,df['Triceps skin fold thickness (mm)'].mean())
df['2-Hour serum insulin (mu U/ml)']=df['2-Hour serum insulin (mu U/ml)'].replace(0,df['2-Hour serum insulin (mu U/ml)'].mean())
df['Body mass index (weight in kg/(height in m)^2)']=df['Body mass index (weight in kg/(height in m)^2)'].replace(0,df['Body mass index (weight in kg/(height in m)^2)'].mean())
# now data makes more sense as it don't contain meaningless zeros
```

```python
df.describe()
```

```python
# independent variables in x and target in y
x = df.drop('Class variable',axis =1 ) #data without Class variable
y= df['Class variable']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
#scikit-learn pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier

from sklearn.pipeline import Pipeline
```

```python
pipeline_lr = Pipeline([('scalar1', StandardScaler()),
    ('lr_classifier', LogisticRegression())
                        ])
#StandardScaler() to standerdize the data in the same format
#LogisticRegression() for classification

pipeline_knn = Pipeline([('scalar2',StandardScaler()),
                         ('knn_classifier',KNeighborsClassifier())])

pipeline_svc = Pipeline([('scalar3',StandardScaler()),
                         ('svc_classifier',SVC())])

pipeline_dt = Pipeline ([('dt_classifier',DecisionTreeClassifier())])
pipeline_rf = Pipeline ([('rf_classifier',RandomForestClassifier(max_depth =3 ))])
pipeline_gbc = Pipeline ([('gbc_classifier',GradientBoostingClassifier())])
#knn Scales the features, then applies K-Nearest Neighbors classification.
#svc Scales the features, then applies a Support Vector Classifier.
```

```python
pipelines = [
            pipeline_lr,
            pipeline_knn,
            pipeline_svc,
            pipeline_dt,
            pipeline_rf,
            pipeline_gbc]
```

```python
pipelines
```

```python
for pipe in pipelines:
    pipe.fit(x_train,y_train)
    # to train all of them
```

```python
pipe_dict = { 0:'LR',1:'KNN',2:'SVC' , 3:'DT', 4:'RF', 5:'GBC'}
```

```python
pipe_dict
```

```python
for i , model in enumerate(pipelines):
    print("{} test Accuracy: {}".format(pipe_dict[i],model.score(x_test,y_test)*100))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'poly', 'rbf', 'sigmoid']
}

grid = GridSearchCV(SVC(), param_grid, cv=5, scoring='accuracy')
grid.fit(x_train, y_train)

print("Best parameters:", grid.best_params_)
print("Best CV accuracy:", grid.best_score_)
```
